<b>Different ways to visualize univariate distributions.

```python
# After my research I found that different ways to plot univariate distributions is to use 
#  Box plots
#  Point plots
#  Count plots
#  Histograms
#  Violin plots
plt.figure(figsize=(15,9))
sns.set(style="ticks")

plt.subplot(2,2,1)
p1 = sns.distplot(data['Polio'],color='r')
plt.subplot(2,2,2)
p2 = sns.countplot(data=data,x='Polio')
plt.subplot(2,2,3)
p3 = sns.boxplot(data=data,x='Polio')
plt.subplot(2,2,4)
p4 = sns.violinplot(y=data['Polio'])
plt.show()
```

```python
# Labels and title are not added intentionally
plt.figure(figsize=(18,18))
sns.set(style='whitegrid')
plt.subplot(3,2,1)
plot1 = sns.pointplot(data=data,x=data.Country,y=data['AdultMortality'])
plt.subplot(3,2,2)
plot2 = sns.barplot(data=data,x=data.Country,y=data['AdultMortality'],hue='Status')
plt.subplot(3,2,3)
plot3 = sns.boxplot(data=data,x=data.Country,y=data['AdultMortality'],hue='Status')
plt.subplot(3,2,4)
plot4 = sns.violinplot(data=data,x=data.Country,y=data['AdultMortality'],hue='Status')
plt.subplot(3,2,5)
plot5 = sns.stripplot(data=data,y=data.Country,x=data['AdultMortality'],hue='Status',dodge=True,jitter=True,alpha=1,zorder=1)
sns.pointplot(x=data['AdultMortality'],y=data.Country,hue='Status',data=data,dodge=.532,join=False,palette='dark',scale=.75,ci=None)
plt.show()
```

```python
final = df[df['Country'].isin(['Norway','Italy','Singapore','Japan','Poland','Palau','Mexico','France'])]
final.dropna(inplace=True)
final.isnull().sum()
```

```python
# Using FacetGrid
sns.set(style='darkgrid')
g = sns.FacetGrid(final,col='Country',col_wrap=3)
g.map(sns.distplot,"Lifeexpectancy",hist=False)
plt.show()
```

```python
# Using Subplots
dims = (15,4)
fig, axs = plt.subplots(ncols=4,figsize=dims)
sns.distplot(final.Alcohol,ax=axs[0],color='r')
sns.distplot(final.AdultMortality,ax=axs[1],color='g')
sns.distplot(final.GDP,ax=axs[2])
sns.distplot(final.Population,ax=axs[3],color='orange')
plt.show()
```

<b>Find a dataset with at least four continuous variables and one categorical variable. Create one master plot that gives insight into the variables and their interrelationships, including:<br/>
Probability distributions<br/>
Bivariate relationships<br/>
Whether the distributions or the relationships vary across groups

```python
sns.set(style='dark',color_codes=True)
dims = (15,8)
fig, axs = plt.subplots(2, 4,figsize=dims)
plt.subplots_adjust(left=0.01, bottom=0.01, right=0.99, top=0.99, wspace=None, hspace=None)
sns.distplot(final.Alcohol,ax=axs[0,0],color='r')
sns.distplot(final.AdultMortality,ax=axs[0,1],color='g')
sns.distplot(final.GDP,ax=axs[0,2])
sns.distplot(final.Population,ax=axs[0,3],color='orange')
sns.regplot(x=final.Alcohol,y=final.AdultMortality,ax=axs[1,0],data=final)
sns.barplot(x=final.Country,y=final.AdultMortality,ax=axs[1,1],data=final,hue='Status')
sns.boxplot(x=final.Country,y=final.GDP,ax=axs[1,2],data=final,hue='Status')
sns.barplot(y=final.Country,x=final.Population,ax=axs[1,3],data=final,hue='Status')
plt.show()
```
